```python
from circles import hough_circle_fixed_radius
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from course_layer import detect_course_impl
```

```python
img = cv2.imread('data/251_crop.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
course = detect_course_impl(img)
plt.imshow(course, cmap='gray')
```

```python
course.max()
```

```python
circles = hough_circle_fixed_radius(course, 31, 700, 750, visualize=False)
```

```python
from circles import circle_mask
n_res = np.zeros(course.shape)
for circle in circles:
    n_res += circle_mask((circle[0],circle[1],31), course.shape).reshape((course.shape[0], course.shape[1]))
plt.imshow(n_res + course / 255, cmap='gray')
plt.show()
```
